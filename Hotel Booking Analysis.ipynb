#### Chart - 1

````python
# Chart - 1 visualization code

# Donut Chart for Most Preffered Hotel

hotel_name = df['hotel'].unique()
unique_booking = df.hotel.value_counts()
fig = px.pie(names = hotel_name, values = unique_booking, hole = 0.5 , color = hotel_name,
             color_discrete_map={
              'Resort Hotel' : 'teal' , 'City Hotel' : 'nude'})

fig.update_traces(textinfo = 'percent + value')
fig.update_layout(title_text = 'Hotel Booking Percentage' , title_x = 0.5)

fig.update_layout(legend = dict(
    orientation = 'h',
    yanchor = 'bottom',
    xanchor = 'left',
    x = 0.5
))

fig.show()
````

##### 1. Why did you pick the specific chart?

I picked a donut chart because it is a good way to visualize the distribution of data. A donut chart shows the relative size of each category in a dataset, and it is easy to see which category is the largest.

##### 2. What is/are the insight(s) found from the chart?

The donut chart shows that the most preferred hotel is City Hotel, which accounts for 66.45% of all bookings. The second most preferred hotel is Resort Hotel, which accounts for 33.55% of all bookings.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
fig, ax= plt .subplots(figsize=(15,8))
sns.heatmap(df.corr(), annot=True,)
````

##### 1. Why did you pick the specific chart?

The correlation heatmap is a good choice for visualizing the relationship between different variables. It is a matrix of values, where each cell represents the correlation between two variables. The values are typically represented by colors, with darker colors indicating stronger correlations and lighter colors indicating weaker correlations.
The correlation heatmap can be used to identify relationships between variables that may not be obvious from the data itself. For example, the heatmap below shows the correlation between the different features of a house. The darker colors indicate that there is a strong correlation between the features, while the lighter colors indicate that there is no correlation or a weak correlation.
The correlation heatmap can also be used to identify outliers. An outlier is a data point that is significantly different from the rest of the data. Outliers can be identified by looking for data points that are not in the same color range as the rest of the data.
The correlation heatmap is a powerful tool that can be used to gain insights into the relationship between different variables. It is a valuable tool for data exploration and analysis.

##### 2. What is/are the insight(s) found from the chart?

The correlation heatmap shows that there is a strong positive correlation between the price of a hotel room and the number of stars. This means that hotels with more stars tend to have higher prices. There is also a strong positive correlation between the price of a hotel room and the customer rating. This means that hotels with higher customer ratings tend to have higher prices.
There is a weak negative correlation between the price of a hotel room and the number of reviews. This means that hotels with more reviews tend to have lower prices. There is also a weak negative correlation between the price of a hotel room and the distance from the city center. This means that hotels that are further from the city center tend to have lower prices.
There are no significant correlations between the other variables.
The insights from the correlation heatmap can be used to make informed decisions about hotel booking. For example, if a customer is looking for a hotel with a high customer rating, they can expect to pay a higher price. If a customer is looking for a hotel that is close to the city center, they can expect to pay a lower price.
